```python
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# increase display size to enable viewing of all data columns

pd.options.display.max_columns = 40
```

```python
# read in the clean flights csv

flights2018_df = pd.read_csv('flights2018clean.csv', index_col=0)
flights2018_df
```

```python
# create a dataframe of the flights originating in Burbank

burbank2018_df = flights2018_df[(flights2018_df['Origin'] == 'BUR')]
burbank2018_df
```

```python
# read in the burbank 2017 file to be used to compare data points

burbank2017_df = pd.read_csv('burbank2017comp.csv')
burbank2017_df.drop(['Unnamed: 0'], axis=1, inplace=True)
burbank2017_df
```

```python
# check to see whether 2017 and 2018 have the same number of unique destinations

print('2017: ' + str(burbank2017_df['Dest'].nunique()))
print('2018: ' + str(burbank2018_df['Dest'].nunique()))
print('Difference: ' + str(burbank2018_df['Dest'].nunique() - burbank2017_df['Dest'].nunique()))
```

```python
# create a new dataframe with the 2017 and 2018 destinations

comparison_df = pd.DataFrame()
comparison_df['Dest18'] = burbank2018_df['Dest']
comparison_df.reset_index(drop=True, inplace=True)
comparison_df['Dest17'] = burbank2017_df['Dest']
comparison_df
```

```python
# check to see which destinations exist in 2017, but not 2018

comparison_df['Unique'] = comparison_df['Dest17'][~comparison_df['Dest17'].isin(comparison_df['Dest18'])].drop_duplicates()
display(comparison_df['Unique'].nunique())

difference_df = pd.DataFrame()
difference_df = comparison_df['Unique'].value_counts()
difference = difference_df.index.to_list()
difference
```

```python
# check to see which destinations exist in 2018, but not 2017

comparison_df['Unique'] = comparison_df['Dest18'][~comparison_df['Dest18'].isin(comparison_df['Dest17'])].drop_duplicates()
display(comparison_df['Unique'].nunique())

difference_df = pd.DataFrame(columns=['Destination', 'Occurence'])
difference_df = comparison_df['Unique'].value_counts()
difference = difference_df.index.to_list()
difference
```

```python
# drop any rows pertaining to HOU, MDW or BOS

burbank2018_df = burbank2018_df.drop(burbank2018_df[(burbank2018_df['Dest'] == 'BOS')].index)
burbank2018_df = burbank2018_df.drop(burbank2018_df[(burbank2018_df['Dest'] == 'MDW')].index)
burbank2018_df = burbank2018_df.drop(burbank2018_df[(burbank2018_df['Dest'] == 'HOU')].index)
burbank2018_df['Dest'].value_counts()
```

```python
# check to see whether 2017 and 2018 have the same number of unique reporting airlines

print('2017: ' + str(burbank2017_df['Reporting_Airline'].nunique()))
print('2018: ' + str(burbank2018_df['Reporting_Airline'].nunique()))
print('Difference: ' + str(burbank2018_df['Reporting_Airline'].nunique() - burbank2017_df['Reporting_Airline'].nunique()))
```

```python
# create a new dataframe with the 2017 and 2018 reporting airlines

comparison_df = pd.DataFrame()
comparison_df['Airline18'] = burbank2018_df['Reporting_Airline']
comparison_df.reset_index(drop=True, inplace=True)
comparison_df['Airline17'] = burbank2017_df['Reporting_Airline']
comparison_df
```

```python
# check to see which reporting airlines exist in 2017, but not 2018

comparison_df['Unique'] = comparison_df['Airline17'][~comparison_df['Airline17'].isin(comparison_df['Airline18'])].drop_duplicates()
display(comparison_df['Unique'].nunique())

difference_df = pd.DataFrame()
difference_df = comparison_df['Unique'].value_counts()
difference = difference_df.index.to_list()
difference
```

```python
# check to see which destinations exist in 2018, but not 2017

comparison_df['Unique'] = comparison_df['Airline18'][~comparison_df['Airline18'].isin(comparison_df['Airline17'])].drop_duplicates()
display(comparison_df['Unique'].nunique())

difference_df = pd.DataFrame(columns=['Destination', 'Occurence'])
difference_df = comparison_df['Unique'].value_counts()
difference = difference_df.index.to_list()
difference
```

```python
# drop any rows pertaining to HOU, MDW or BOSburbank2018_df = burbank2018_df.drop(burbank2018_df[(burbank2018_df['Reporting_Airline'] == 'YV')].index)

burbank2018_df = burbank2018_df.drop(burbank2018_df[(burbank2018_df['Reporting_Airline'] == 'YV')].index)
burbank2018_df['Reporting_Airline'].value_counts()
```

```python
# drop columns with redundant, 'give-away', or unuseful information

burbank2018_df.drop(['Tail_Number', 
                     'OriginCityName', 
                     'DestCityName',
                     'DepTime', 
                     'DepDel15', 
                     'DepartureDelayGroups', 
                     'TaxiOut', 
                     'WheelsOff',
                     'DistanceGroup',
                     'CarrierDelay',
                     'WeatherDelay',
                     'NASDelay',
                     'SecurityDelay',
                     'LateAircraftDelay',
                     'CancellationCode'], inplace=True, axis=1)
burbank2018_df
```

```python
# import all the weather data

weather_df = pd.read_csv('data/weather/all13_weather.csv')
weather_df
```

```python
# create a dataframe of the 2018 weather

weather2018_df = weather_df[(pd.DatetimeIndex(weather_df['DATE']).year == 2018)]
weather2018_df
```

```python
# look at the total number of NaN values for each column

weather2018_df.isna().sum()
```

```python
# TMAX and TMIN will be used to generate missing values for TAVG and are only missing a few values -- fill using forward fill

weather2018_df['TMAX'].ffill(inplace=True)
weather2018_df['TMIN'].ffill(inplace=True)
weather2018_df.isna().sum()
```

```python
# create a temporary column calculating the average temperature using TMAX and TMIN

weather2018_df['temp_avg'] = (weather2018_df['TMAX'] + weather2018_df['TMIN']) / 2
weather2018_df
```

```python
# fill in the missing TAVG values using the newly created average temperature column

weather2018_df['TAVG'].fillna(weather2018_df['temp_avg'], inplace=True)
weather2018_df.isna().sum()
```

```python
# drop unnecessary and/or redundant columns

weather2018_df.drop(['STATION', 
                     'NAME', 
                     'PGTM', 
                     'SNWD', 
                     'TMAX', 
                     'TMIN', 
                     'TOBS', 
                     'WDF2', 
                     'WDF5', 
                     'WSF2', 
                     'WSF5',
                     'temp_avg'], axis=1, inplace=True)
weather2018_df
```

```python
# check for any remaining NaN values

weather2018_df.isna().sum()
```

```python
# investigate which airports are missing values for SNOW

weather2018_df['AIRPORT'][(weather2018_df['SNOW'].isna())].value_counts()
```

```python
# all airports in warm climates where snow is unlikely... so input zeros for the NaNs

weather2018_df['SNOW'].fillna(0, inplace=True)
weather2018_df.isna().sum()
```

```python
# investigate the row missing precipitation information

weather2018_df[(weather2018_df['PRCP'].isna())]
```

```python
# fill in these missing values using forward fill

weather2018_df['PRCP'].ffill(inplace=True)
weather2018_df.isna().sum()
```

```python
# rename the columns with more intuitive names

weather2018_df.columns = ['AIRPORT', 
                          'DATE', 
                          'WIND', 
                          'PRCP', 
                          'SNOW', 
                          'TEMP']
weather2018_df
```

```python
# ensure that the date column in both datafranes is of type datetime

pd.to_datetime(burbank2018_df['FlightDate'])
pd.to_datetime(weather2018_df['DATE'])
```

```python
# ensure there is an entire year of data

display(burbank2018_df['FlightDate'].min())
display(burbank2018_df['FlightDate'].max())
```

```python
# add in the origin airport weather to the burbank dataframe

burbank2018_df = pd.merge(burbank2018_df, weather2018_df, left_on=['FlightDate', 'Origin'], right_on=['DATE', 'AIRPORT'], how='left')
burbank2018_df.drop(columns=['AIRPORT', 'DATE'], axis=1, inplace=True)
burbank2018_df
```

```python
# ensure the start and end dates are still correct

display(burbank2018_df['FlightDate'].min())
display(burbank2018_df['FlightDate'].max())
```

```python
# check that the new columns have the correct values for Jan 1, 2018

burbank2018_df[(burbank2018_df['FlightDate'] == '2018-01-01')].head()
```

```python
# rename the new weather columns so that we know they pertain to the origin airport

burbank2018_df.rename(columns={'WIND' : 'Origin_WIND',
                                'PRCP' : 'Origin_PRCP',
                                'SNOW' : 'Origin_SNOW',
                                'TEMP' : 'Origin_TEMP'}, inplace=True)
burbank2018_df.head()
```

```python
# add in the destination airport weather

burbank2018_df = pd.merge(burbank2018_df, weather2018_df, left_on=['FlightDate', 'Dest'], right_on=['DATE', 'AIRPORT'], how='left')
burbank2018_df.drop(columns=['AIRPORT', 'DATE'], axis=1, inplace=True)
burbank2018_df
```

```python
# rename the new weather columns so that we know they pertain to the destination airport

burbank2018_df.rename(columns={'WIND' : 'Dest_WIND',
                               'PRCP' : 'Dest_PRCP',
                               'SNOW' : 'Dest_SNOW',
                               'TEMP' : 'Dest_TEMP'}, inplace=True)                               
burbank2018_df
```

```python
# ensure there are no NaN values in the burbank dataframe

burbank2018_df.isna().sum()
```

```python
# set the Date to be a datecolumn and extract each of day, month and year separately, create own column for each
# drop the orignal Date column and check that the new columns are correct

burbank2018_df['Day'] = pd.DatetimeIndex(burbank2018_df['FlightDate']).day
burbank2018_df['Month'] = pd.DatetimeIndex(burbank2018_df['FlightDate']).month
burbank2018_df['Year'] = pd.DatetimeIndex(burbank2018_df['FlightDate']).year
burbank2018_df.drop(['FlightDate'], axis=1, inplace=True)
burbank2018_df
```

```python
# create dummy variables for the airlines

airline_df = pd.get_dummies(burbank2018_df['Reporting_Airline'], drop_first=True)
airline_df
```

```python
# create dummy variables for the destination airports

dest_df = pd.get_dummies(burbank2018_df['Dest'], drop_first=True)
dest_df = dest_df.add_prefix('da_')
dest_df
```

```python
# combine the dummy variables and the original dataframe

burbank2018_df = pd.concat([burbank2018_df, airline_df, dest_df], axis=1) 
burbank2018_df.drop(['Reporting_Airline', 'Dest'], inplace=True, axis=1)
burbank2018_df
```

```python
# move the delayed (target) column to the end

delay_df = burbank2018_df.pop('Delayed')
burbank2018_df['Delayed'] = delay_df
burbank2018_df
```

```python
# look at the proportion of delayed flights

burbank2018_df['Delayed'].value_counts()
```

```python
# check that all values are numerical (float or int)

burbank2018_df.dtypes
```

```python
# drop the Year. Origin and OriginState columns because they only have one value each
# drop information about actual arrival times (gives away too much information)

burbank2018_df.drop(columns=['Year',
                             'Origin',
                             'OriginState',
                            #'DepDelay',
                             'ArrTime',
                             'ArrDelay',
                             'DestState'], axis=1, inplace=True)
burbank2018_df 
```
