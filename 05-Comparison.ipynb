**Comparison**
<br/>Compare to see which airlines exist in 2017, but not 2018 and vice versa.

```python
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# increase display size to enable viewing of all data columns

pd.options.display.max_columns = 35
pd.options.display.max_rows = 360
```

```python
# read in the 2017 flights csv

flights2017_df = pd.read_csv('flights2017clean.csv', index_col=0)
flights2017_df
```

```python
# read in the 2018 flights csv

flights2018_df = pd.read_csv('flights2018clean.csv', index_col=0)
flights2018_df
```

```python
# check to see whether 2017 and 2018 have the same number of unique airports

print('2017: ' + str(flights2017_df['Origin'].nunique()))
print('2018: ' + str(flights2018_df['Origin'].nunique()))
print('Difference: ' + str(flights2018_df['Origin'].nunique() - flights2017_df['Origin'].nunique()))
```

```python
# create a new dataframe with the 2017 and 2018 origin airports

comparison_df = pd.DataFrame()
comparison_df['Airlines18'] = flights2018_df['Origin']
comparison_df['Airlines17'] = flights2017_df['Origin']
comparison_df
```

```python
# check to see which airports exist in 2017, but not 2018

comparison_df['Unique'] = comparison_df['Airlines17'][~comparison_df['Airlines17'].isin(comparison_df['Airlines18'])].drop_duplicates()
comparison_df
```

```python
# create a list if the non-duplicated values

display(comparison_df['Unique'].nunique())

difference_df = pd.DataFrame()
difference_df = comparison_df['Unique'].value_counts()
difference = difference_df.index.to_list()
difference
```

```python
# check to see which dataframe the non-duplicated values are in

print('2017: ' + str(flights2017_df['Origin'].loc[flights2017_df['Origin'].isin(difference)].nunique()))
print('2018: ' + str(flights2018_df['Origin'].loc[flights2018_df['Origin'].isin(difference)].nunique()))
```

```python
# check to see which airports exist in 2018, but not 2017

comparison_df['Unique'] = comparison_df['Airlines18'][~comparison_df['Airlines18'].isin(comparison_df['Airlines17'])].drop_duplicates()
comparison_df
```

```python
# create a list if the non-duplicated values and create a csv to be used in other notebooks

display(comparison_df['Unique'].nunique())

difference_df = pd.DataFrame(columns=['Airline', 'Occurence'])
difference_df = comparison_df['Unique'].value_counts()

difference = difference_df.index.to_list()
difference
```

```python
# check to see which dataframe the non-duplicated values are in

print('2017: ' + str(flights2017_df['Origin'].loc[flights2017_df['Origin'].isin(difference)].nunique()))
print('2018: ' + str(flights2018_df['Origin'].loc[flights2018_df['Origin'].isin(difference)].nunique()))
```

Based on these results, it seems likely there are large differences in the unique airlines and destinations between the three years. <br/>Further comparisons were not made at this time.
